# Question 4

```python
import cv2
```

```python
pic = cv2.imread('11.jpg', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(20,10))
plt.subplot(1,4,1)
plt.imshow(pic, cmap ='gray')
plt.title('Original')

vertical_filter = np.array(
[[1, 0, -1],
[1, 0, -1],
[1, 0, -1]]
)

pic2 = cv2.filter2D(pic, -1, vertical_filter)
plt.subplot(1,4,2)
plt.imshow(pic2, cmap ='gray')
plt.title('vertical_filter')

horizontal_filter = np.array(
[[1, 1, 1],
[0, 0, 0],
[-1, -1, -1]]
)

pic3 = cv2.filter2D(pic, -1, horizontal_filter)
plt.subplot(1,4,3)
plt.imshow(pic3, cmap ='gray')
plt.title('horizontal_filter')

gaussian_filter = np.array(
[
    [1,2,3,2,1],
    [2,5,6,5,2],
    [3,6,8,6,3],
    [2,5,6,5,2],
    [1,2,3,2,1]
])
gaussian_filter = gaussian_filter/np.sum(gaussian_filter)
pic4 = cv2.filter2D(pic, -1, gaussian_filter)
plt.subplot(1,4,4)
plt.imshow(pic4, cmap ='gray')
plt.title('gaussian_filter')

plt.show()
```
